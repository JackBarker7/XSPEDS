```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ones = np.ones((2000,2000))


def create_arc_mask(ones, radius, center, width):
    """Create an arc mask for a given arc radius, center (in format (x,y)), and width."""
    center = center[::-1]
    x, y = np.indices((ones.shape))
    r = np.sqrt((x - center[0])**2 + (y - center[1])**2)
    mask = np.logical_and(r > radius - width/2, r < radius + width/2)
    return mask
```

```python
plt.imshow(create_arc_mask(ones, 6663, (8080,856), 10))
```
