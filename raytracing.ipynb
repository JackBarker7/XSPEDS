```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=10

o = [0,0,a]
theta = np.pi/4

beta = np.linspace(0,2*np.pi,100)

# find the points of a circle, produced by all vectors of length a and angle theta from o
r = a/np.sin(theta)
x_circ = r*np.cos(beta)
y_circ = r*np.sin(beta)

circle_vectors = np.array([x_circ,y_circ,-a*np.ones(len(x_circ))]).T


# get the reflection of these vectors off the plane z=0
# the normal vector of the plane is [0,0,1]
normal = np.array([0,0,1])

reflected_vectors = np.array([circle_vector - 2*np.dot(circle_vector,normal)*normal for circle_vector in circle_vectors])

reflected_vectors

```
